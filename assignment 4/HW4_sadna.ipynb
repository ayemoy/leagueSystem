```python
#scaling ot handle highly varying magnitudes or values or units.
from sklearn.preprocessing import StandardScaler
def scaling_data(X_train,X_test):

  scaler = StandardScaler()
  scaler.fit(X_train)

  X_train = scaler.transform(X_train)
  X_test = scaler.transform(X_test)
  return X_train,X_test
```

```python
# train the KNN algorithm to make predictions with it
import time
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.neural_network import MLPClassifier


def train_algorithm(X_train, y_train,model):


  classifier = model
  #measure time
  start = time.time()
  
  #fit
  trained_model = classifier.fit(X_train, y_train)
  
  stop = time.time()
  elapsed = stop - start

  return elapsed,trained_model
```

```python
#create empty df to save all the results
measurments = ['Avg Acuracy','Acuracy STD','Avg precision','precision STD','Avg recall','recall STD','Avg f1 score','f1 score STD','Avg auc roc curve score','roc score STD','TPR','FPR','avg inference time','inference STD','train time']
col_names = ['KNN', 'Random forest','Decision tree' , 'ANN']
saved_df = pd.DataFrame(columns=['measurment']+col_names)
for i in range(len(measurments)):
    saved_df.loc[i] = [measurments[i]] + ['Nan']*4
saved_df
```

```python
#function to calculate precision recall

def calculate_precision_recall(y_test_, _y_pred_,precision_recall_curve_list,roc_curve_list,auc_roc_curve_list): 

  #measure presicion and recall for the curve
  precision, recall, thresholds = precision_recall_curve(list(map(int, y_test_)),list(map(int, _y_pred_)))
  precision_recall_curve_list.append([precision,recall])

  #get fpr,tpr for the curve
  fpr, tpr, _ = metrics.roc_curve(list(map(int, y_test_)),list(map(int, _y_pred_)))
  
  roc_curve_list.append([fpr,tpr])
  auc = metrics.roc_auc_score(y_test_,  _y_pred_)
  
  auc = float("{:.7f}".format(auc))
  auc_roc_curve_list.append(auc)

```

```python
#transform multi vector to single vector in order to plot a auc roc curve
def binary_transformer(milti_class_vector):
        
        #cahnge to int
        milti_class_vector = [ int(x) for x in milti_class_vector ]

        #transform to binary
        transformer = Binarizer().fit([milti_class_vector])
        binary_vector= transformer.transform([milti_class_vector])
        binary_vector = binary_vector.tolist()
        binary_vector =binary_vector[0]
        binary_vector = [ int(x) for x in binary_vector ]

        #transform back to df
        binary_vector = pd.Series(binary_vector)
        return binary_vector
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure


#plot auc curve

def plot_roc_curve(roc_curve_list,k,auc_roc_curve_list,name):
    auc = sum(auc_roc_curve_list)/k

    # k = 10 because 10 cross validation
    for i in range(k):
      fpr = (roc_curve_list[i][0])
      tpr = (roc_curve_list[i][1])
      plt.plot(fpr,tpr,label="K in k-fold="+str(i)) 

    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.legend(loc=4)
    plt.title(name)

    plt.show()
```

```python
import statistics
from sklearn.metrics import f1_score
from sklearn.model_selection import KFold 
from sklearn.metrics import recall_score
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import roc_curve
from sklearn.preprocessing import Binarizer
import matplotlib.pyplot as plt


#fuknction to calculate all measurements

def calculate_ML_measurements(model,name,col_name,space):


  #10 fold cross validation
  k = 10
  kf = KFold(n_splits=k, random_state=None)


  #list to save all resualts
  acc_score = []
  prec_score = []
  recc_score = []
  f1_Score = []
  time_list = []
  auc_roc_curve_list = []
  TPR_list = []
  FPR_list = []
  precision_recall_curve_list = []
  roc_curve_list = []
  one_measure_of_each = []
  inference_time_list = []

  
  #preform 10 folds
  for train_index , test_index in kf.split(X):
      X_train , X_test = X.iloc[train_index,:],X.iloc[test_index,:]
      y_train , y_test = Y[train_index] , Y[test_index]
     
     
      #scale the data
      X_train,X_test = scaling_data(X_train,X_test)


      #find the best paramaters
      par = calculate_hyperparamaters(model,space,name,X_train,y_train)
      model.set_params(**par)



      #learn and measure the time
      elapsed_time,model = train_algorithm(X_train,y_train,model)
      time_list.append(elapsed_time)
      
      #make the prediction
      start_inference = time.time()
      y_pred = model.predict(X_test)
      
      #calculate inference time
      stop_inference = time.time()
      inference_time_list.append(stop_inference-start_inference)

      #if we have more then 2 classes to predict from 
      if len(Y.unique()) >2:

        #transform the test and the prediction from multicalss [0,1,2,3,4....] to binary [0,1]
        binary_y_test = binary_transformer(y_test)
        binary_y_pred=  binary_transformer(y_pred)
        
        calculate_precision_recall( binary_y_test,  binary_y_pred,precision_recall_curve_list,roc_curve_list,auc_roc_curve_list)
      
      #if target class if binary
      else:
        calculate_precision_recall( y_test,  y_pred,precision_recall_curve_list,roc_curve_list,auc_roc_curve_list)
      
      #create a confusion matrix
      cnf_matrix=confusion_matrix(y_test, y_pred)
      
      #extract each result from the matrix
      TP = cnf_matrix[0][0] 
      FN = cnf_matrix[0][1]
      FP = cnf_matrix[1][0]
      TN = cnf_matrix[1][1]
      
      #accuracy
      acc = (TP+TN)/(TP+TN+FP+FN)
      acc = float("{:.7f}".format(acc))
      acc_score.append(acc)
      
      
      #precision
      prec = TP / (TP+FP)
      prec = float("{:.7f}".format(prec))
      prec_score.append(prec)

      #recall
      rec = TP/(TP + FN)
      rec = float("{:.7f}".format(rec))
      recc_score.append(rec)
      
      #f1 score
      f1 = f1_score(y_test,y_pred, average='weighted')
      f1 = float("{:.7f}".format(f1))
      f1_Score.append(f1)

      # False negative rate
      FPR = FP/(FP+TN)
      FPR = float("{:.7f}".format(FPR))
      #True positive rate
      TPR = TP/(TP+FN)
      TPR = float("{:.7f}".format(TPR))

      TPR_list.append(TPR)
      FPR_list.append(FPR)

  #calculate avrereges 
  avg_acc_score = sum(acc_score)/k
  avg_prec_score = sum(prec_score)/k
  avg_recc_score = sum(recc_score)/k
  avg_f1_score = sum(f1_Score)/k
  avg_auc_score = sum(auc_roc_curve_list)/k
  avg_tpr = sum(TPR_list)/k
  avg_fpr = sum(FPR_list)/k

  #inference time
  inference_time = 1000*sum(inference_time_list)/k
  inference_time = inference_time/df.shape[0]

  measurments =measurments = ['Avg Acuracy','Acuracy STD','Avg precision','precision STD','Avg recall','recall STD','Avg f1 score','f1 score STD',
                              'Avg auc roc curve score','roc score STD','TPR','FPR','avg inference time','inference STD','train time']

  #save all averages and their STD
  one_measure_of_each.extend([avg_acc_score,statistics.stdev(acc_score),avg_prec_score,statistics.stdev(prec_score)
                              ,avg_recc_score,statistics.stdev(recc_score),avg_f1_score,statistics.stdev(f1_Score),avg_auc_score,
                              statistics.stdev(auc_roc_curve_list),avg_tpr,avg_fpr,inference_time,statistics.stdev(inference_time_list),sum(time_list)])
  
  #save results to the new df
  for i in range(len(measurments)):
    saved_df._set_value(i,col_name,one_measure_of_each[i])


  #plot the auc curve
  plot_roc_curve(roc_curve_list,k,auc_roc_curve_list,name)

  #prints
  print("outcomes for using: ", name ," classifier")
  print('-------Accuracy-------')
  print('accuracy of each fold - {}'.format(acc_score))
  print('Avg accuracy : {}'.format(avg_acc_score))
  print('----------------------')
  print('-------Precision-------')
  print('precision of each fold - {}'.format(prec_score))
  print('Avg precision : {}'.format(avg_prec_score))
  print('----------------------')
  print('-------Recall-------')
  print('recall of each fold - {}'.format(recc_score))
  print('Avg recall : {}'.format(avg_recc_score))
  print('----------------------')
  print('-------f1 score-------')
  print('f1 score each fold - {}'.format(f1_Score))
  print('Avg f1 score : {}'.format(avg_f1_score))
  print('----------------------')
  print('-------auc-------')
  print('auc roc curve each fold - {}'.format(auc_roc_curve_list))
  print('Avg auc roc curve score : {}'.format(avg_auc_score))
  print('----------------------')
  print('-------TPR AND FPR-------')
  print('True positive ratio for each fold - {}'.format(TPR_list))
  print('Avg True positive ratio : {}'.format(avg_tpr))
  print('----------------------')
  print('False positive ratio for each fold - {}'.format(FPR_list))
  print('Avg False positive ratio : {}'.format(avg_fpr))
  print('----------------------')
  print('-------Time-------')
  print('time for each fold in seconds: {}'.format(time_list))
  print('total run time in seconds: {}'.format(sum(time_list)))
```

```python
# random search model
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import RandomizedSearchCV

def calculate_hyperparamaters(model,space,name,x_,y_):
  
  # define evaluation
  cv = RepeatedStratifiedKFold(n_splits=3, n_repeats=3, random_state=1)
  # define search
  search = RandomizedSearchCV(model, space, n_jobs=-1, cv=cv, random_state=1)
  # execute search
  result = search.fit(x_, y_)
  # summarize result
  print(f'Best Hyperparameters for {name} are:  { result.best_params_}')
  return result.best_params_
```
