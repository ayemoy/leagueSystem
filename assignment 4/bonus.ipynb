```python

import pandas as pd
import matplotlib.pyplot as plt

import numpy as np
import scikit_posthocs as sp
from scipy.stats import wilcoxon, friedmanchisquare, rankdata
url = '/content/sample_data/all-results.csv'
df = pd.read_csv(url)
```

```python
df
```

```python
# array of accuarcy
knn_array = df['KNN'].to_numpy()
random_forest_array = df['Random forest'].to_numpy()
Decision_tree_array = df['Decision tree'].to_numpy()
ANN_array = df['ANN'].to_numpy()
# array of time
knn_array_time = df['KNN-time'].to_numpy()
random_forest_array_time = df['Random forest-time'].to_numpy()
Decision_tree_array_time = df['Decision tree- time'].to_numpy()
ANN_array_time = df['ANN-time'].to_numpy()


res_acc=friedmanchisquare(knn_array,random_forest_array,Decision_tree_array,ANN_array)
res_time=friedmanchisquare(knn_array_time,random_forest_array_time,Decision_tree_array_time,ANN_array_time)
```

```python
if res_acc[1] <= 0.05:
    dataframe_acc=sp.posthoc_nemenyi_friedman(sp.posthoc_nemenyi_friedman(np.array([knn_array,random_forest_array,Decision_tree_array,ANN_array]).T))
    dataframe_acc.columns = ['KNN', 'Random forest', 'Decision tree', 'ANN']
    dataframe_acc.index= ['KNN', 'Random forest', 'Decision tree', 'ANN']
    print("Friedman Test Accuracy:")
    print(res_acc)
    print("Post hoc Accuracy:")
    print(dataframe_acc, "\n")


if res_time[1] <= 0.05:
    dataframe_time=sp.posthoc_nemenyi_friedman(sp.posthoc_nemenyi_friedman(np.array([knn_array_time,random_forest_array_time,Decision_tree_array_time,ANN_array_time]).T))
    dataframe_time.columns = ['KNN', 'Random forest', 'Decision tree', 'ANN']
    dataframe_time.index= ['KNN', 'Random forest', 'Decision tree', 'ANN']
    print("Friedman Test Time:")
    print(res_time)
    print("Post hoc Time:")
    print(dataframe_time)
```
